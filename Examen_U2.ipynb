# Hipotesis:

Hipótesis nula (H₀):
No existe una relación significativa entre la masa corporal, tensión arterial, peso y medida de cintura con el riesgo de hipertensión.

Hipótesis alternativa (H₁):
Existe al menos una relación significativa entre la masa corporal, tensión arterial, peso y medida de cintura con el riesgo de hipertensión.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import matplotlib.pyplot as plt

X = df[[ 'peso', 'medida_cintura', 'tension_arterial', 'masa_corporal']]
Y = df['riesgo_hipertension']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42 )

x_train_constante = sm.add_constant(X_train)
x_test_constante = sm.add_constant(X_test)

modelo = sm.OLS(Y_train, x_train_constante).fit()

Yc = modelo.predict(x_test_constante)

plt.scatter(Y_test, Yc)

r2 = r2_score(Y_test, Yc)
print(f'Coeficiente de determinación: {r2: 0.4}')

from statsmodels.formula.api import ols

moedo_2 = ols(formula = 'riesgo_hipertension ~ peso + medida_cintura + tension_arterial + masa_corporal', data = df).fit()
tabla_anova = sm.stats.anova_lm(moedo_2, typ = 2)  
tabla_anova
```

```python
b0, b1, b2, b3, b4,  = modelo.params

Fun = lambda X1, X2, X3, X4: b0 + b1 * X1 + b2 * X2 + b3 * X3 + b4 * X4 

#Eres el ecnargado de evaluar a un paciente que tiene las sospechas de ser hipertenso, al hacer un chequeo medico rapido anotas la siguiente informacion ¿Es probable que tu paciente sea hipertenso? .

Fun(103.75, 118, 128, 33.07)
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

stats.probplot(modelo.resid, dist="norm", plot=plt)
plt.title( 'Redisuales')
plt.show()

plt.hist(modelo.resid, bins=30)
plt.title("Histograma de los residuos")
plt.xlabel("Residuo")
plt.ylabel("Frecuencia")
plt.show()
```

```python
from scipy.stats import shapiro

stat, p = shapiro(modelo.resid)
print('Estadístico Shapiro-Wilk=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print("Los residuos parecen normales.")
```

```python
from statsmodels.stats.diagnostic import het_breuschpagan

_, pval, __, ___ = het_breuschpagan(modelo.resid, modelo.model.exog)
print("p-valor de Breusch-Pagan:", pval)
```

# Interpretacion de resultados generales

Con base en los resultados obtenidos del ANOVA, se evaluó el efecto de cuatro variables predictoras —peso, medida de cintura, tensión arterial y masa corporal— sobre la variable dependiente asociada al riesgo de hipertensión. El coeficiente de determinación (R²) obtenido fue de 0.3059, lo cual indica que el modelo explica aproximadamente el 30.59% de la variabilidad de la variable dependiente.

Los resultados muestran que tres de las cuatro variables analizadas presentan una relación estadísticamente significativa con la variable dependiente, al presentar valores-p (p-value) inferiores al umbral convencional de significancia (α = 0.05), específicamente:

Medida de cintura (p = 1.27e-05)

Tensión arterial (p = 2.09e-60)

Masa corporal (p = 4.90e-24)

Por el contrario, la variable peso (p = 0.3146) no mostró un efecto significativo en el modelo.

Dado que al menos una de las variables independientes presenta un efecto significativo, se rechaza la hipótesis nula (H₀). Por tanto, se concluye que existe evidencia estadística suficiente para afirmar que al menos una de las variables analizadas influye significativamente en el riesgo de hipertensión en la población estudiada.

# Interpretacion resultado del modelo

Despues de ingresar los datos del paciente dentro del modelo creado, se puede observar que el paciente en efecto tiene un alto riesgo de ser hipertenso o ya serlp.

# Normalidad

El Shapiro-Wilk prueba la hipótesis nula de que los datos (residuos) siguen una distribución normal.

Como el p-valor es muy bajo (0.000 < 0.05), se rechaza la hipótesis nula.

Conclusión: Los residuos no siguen una distribución normal, lo que viola el supuesto de normalidad.

# Homosedasticidad

Como el p-valor es prácticamente cero, se rechaza la hipótesis nula.

Conclusión: Se detecta heterocedasticidad (los residuos no tienen varianza constante), lo cual viola el supuesto de homoscedasticidad.

# Conclusion general

El análisis estadístico realizado permitió identificar factores significativos asociados al riesgo de hipertensión en la población evaluada. A través del modelo de regresión lineal, se encontró que medida de cintura, tensión arterial y masa corporal son variables predictoras que presentan una relación estadísticamente significativa con el riesgo de hipertensión (p < 0.05), mientras que la variable peso no mostró un efecto relevante dentro del modelo (p > 0.05).

El coeficiente de determinación (R² = 0.3059) indica que el modelo explica aproximadamente el 30.59% de la variabilidad en la variable dependiente, lo cual representa un nivel aceptable en estudios de carácter biomédico o social, donde múltiples factores suelen influir en los resultados.

Sin embargo, al evaluar los supuestos clásicos del modelo, se detectaron violaciones importantes:

Los residuos no presentan distribución normal según la prueba de Shapiro-Wilk (p < 0.05).

Se identificó heterocedasticidad, lo que indica que la varianza de los errores no es constante (p ≈ 0 en la prueba de homocedasticidad).

Estos resultados sugieren que, si bien el modelo logra identificar variables significativas, no cumple con todos los supuestos requeridos para una regresión lineal clásica, por lo que sus estimaciones podrían estar sesgadas o ser ineficientes. Se recomienda complementar el análisis con modelos más robustos o transformaciones que corrijan las desviaciones observadas.

Finalmente, al aplicar el modelo a un caso práctico, los datos de un paciente específico arrojaron una alta probabilidad de hipertensión, lo que respalda la utilidad predictiva del modelo como herramienta de apoyo en la identificación de riesgos clínicos, siempre y cuando se tomen en cuenta sus limitaciones estadísticas.
